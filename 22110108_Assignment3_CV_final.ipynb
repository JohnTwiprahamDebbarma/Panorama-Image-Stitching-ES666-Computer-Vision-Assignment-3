```python
import cv2
import matplotlib.pyplot as plt

for i in range(len(allims)):
  read_image = cv2.imread(allims[i])
  plt.figure()
  plt.axis('off')
  plt.title("Image" + str(i))
  read_image = cv2.cvtColor(read_image, cv2.COLOR_BGR2RGB)
  plt.imshow(read_image)
```
